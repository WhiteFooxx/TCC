```python
import pandas as pd
import json 
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.mlab as mlab
```

```python
with open("tcc-silo-default-rtdb-export.json", encoding='utf-8') as meu_json:
  dados = json.load(meu_json)
```

```python
dadosData = dados['Data']
```

```python
valores = []

valoresTemp = []
valoresHumid = []

for i in dadosData:

  dadosTemp = dadosData[i]['temp']
  dadosHumid = dadosData[i]['humid']  


  if dadosTemp != '' or dadosHumid != '':
    valoresTemp.append(float(dadosTemp))
    valoresHumid.append(float(dadosHumid))
  else:
    continue
```

```python
plt.figure(figsize=(20, 8))
plt.subplot(2, 1, 1)
plt.grid()
plt.title('Dados de Temperatura e Umidade', color='black', fontsize=20)
plt.yticks([15,20,25,30,35,40])
plt.ylabel('Temperatura (Â°C)', fontsize=16)
plt.plot(valoresTemp, 'orange')

plt.subplot(2, 1, 2)
plt.grid()
plt.ylabel('Umidade (%)', fontsize=16)
plt.plot(valoresHumid, 'blue')

plt.tight_layout()
plt.show()
```
